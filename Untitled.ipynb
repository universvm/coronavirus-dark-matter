# The Dark Matter of the novel Coronavirus

Symptomps of COVID19 include high fever, perpetual coughs and lack of taste (#https://www.who.int/emergencies/diseases/novel-coronavirus-2019/question-and-answers-hub/q-a-detail/q-a-coronaviruses#:~:text=symptoms). In Italy, the data is showing that a significant numbers of the cases are actually in the 50+ age group. Why is this? Could it be that the older population is showing stronger symptoms and therefore is likely to be diagnosed earlier? 


To answer this question, I plan to analize the distribution of the population by age and try to estimate the number of people missing from each age group. There are major assumptions with this approach:


1. The virus spreads evenly through the population
2. Age does not affect likelihood of infection, just the severity of the symptoms. 
3. People of different ages have similar behaviours and interact with similar distribution of age groups.

These are large assumptions since in Italy for example, a lot of hospitals have been hit by coronavirus outbreaks early on (source: https://fortune.com/2020/03/19/coronavirus-south-korea-test-data/). Nonetheless, we're after an upper estimate of the number of infected people.  

## How do you know if the number is correct? 

I tried to think of several ways to "cast out the nines". Fist, I thought about the estimation of total cases from the death rate. Ideally, if the numbers I calculate are sensible, the estimation of total cases from the death rate should be relativey similar. Second, I thought about doing something similar for countries where a significant number of tests have been done, eg. South Korea and Germany. If the logic is correct, their coronavirus cases by age should be similar to the population distribution by age. Else, either the assumptions are incorrect, or there is something special about these countries health care system / measures. 

Additionally, the death rate calculated should be approximately the same. 

```python
# Analysys
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

# Set dimension:
dims = (20, 8.27) # Custom styles:
plt.figure(figsize=dims)
plt.style.use('ggplot')
# Inline jupyter:
%matplotlib inline
```

# Population Statistics 

Let's start off with basic stats for the populations. I'll be using the UN as my source:

https://population.un.org/wpp/Download/Standard/Population/

```python
population_stats = pd.read_csv('data/population.csv')
# Select population numbers only:
italy_pop_stats = population_stats[population_stats['Region, subregion, country or area *'] == 'Italy']
germany_pop_stats =  population_stats[population_stats['Region, subregion, country or area *'] == 'Germany']
skorea_pop_stats =  population_stats[population_stats['Region, subregion, country or area *'] == 'RepublicofKorea']
```

```python
# Convert to percentage
def calculate_percentage_in_stats(country_data):
    
    country_name = country_data[country_data.columns[2]]
    # Sum row and divide by sum
    sum_row = country_data[country_data.columns[8:]].sum(axis=1).item()
    country_data = country_data[country_data.columns[8:]].divide(sum_row/100)
    country_data['country'] = country_name.values
    return country_data    

italy_percentage_stats = calculate_percentage_in_stats(italy_pop_stats)
germany_percentage_stats = calculate_percentage_in_stats(germany_pop_stats)
skorea_percentage_stats = calculate_percentage_in_stats(skorea_pop_stats)
```

```python
def create_population_graph(country_data, color_id):
     # Matplotlib object:
    fig, ax = plt.subplots(figsize=dims) 
    # Pass data + Order:
    ax = sns.barplot(data=country_data, color=sns.color_palette("viridis")[color_id])
    # Tight Layout
    plt.tight_layout()
    # Add titles
    country_name = country_data[country_data.columns[-1]].item()
    plt.title(f'{country_name}\'s Percentage Population Distribution by Age') 
    plt.xlabel('Age Group (years)')
    plt.ylabel('Percentage (%)')
```

```python
create_population_graph(italy_percentage_stats, 1)
create_population_graph(germany_percentage_stats, 3)
create_population_graph(skorea_percentage_stats, 5)
```

# Coronavirus Distribution Statistics

Let's now have a look at the distribution of coronavirus cases by age. The source for this data is:

- Italy - https://www.epicentro.iss.it/coronavirus/bollettino/Bollettino-sorveglianza-integrata-COVID-19_26-maggio-2020.pdf
- South Korea - https://www.cdc.go.kr/board/board.es?mid=a20501000000&bid=0015&list_no=367377&act=view

```python
corona_stats = pd.read_csv('data/coronavirus.csv')
# Select population numbers only:
italy_corona_stats = corona_stats[corona_stats['Country'] == 'Italy']
skorea_corona_stats  =  corona_stats[corona_stats['Country'] == 'RepublicofKorea']
```

```python
def create_corona_graph(country_data, color_id):
     # Matplotlib object:
    fig, ax = plt.subplots(figsize=dims) 
    # Pass data + Order:
    coronavirus_data = country_data[country_data.columns[0:]]
    #print(country_data)
    ax = sns.barplot(data=coronavirus_data, color=sns.color_palette("viridis")[color_id])
    # Tight Layout
    plt.tight_layout()
    # Add titles
    country_name = country_data[country_data.columns[-1]].item()
    plt.title(f'{country_name}\'s Coronavirus Distribution by Age') 
    plt.xlabel('Age Group (years)')
    plt.ylabel('Numbers')
```

```python
# Convert to percentage
def calculate_percentage_in_corona(country_data):
    # Sum row and divide by sum
    country_name = country_data[country_data.columns[0]]
    sum_row = country_data[country_data.columns[1:]].sum(axis=1).item()
    country_data = country_data[country_data.columns[1:]].divide(sum_row/100)
    country_data['country'] = country_name.values
    return country_data    

italy_percentage_corona = calculate_percentage_in_corona(italy_corona_stats)
skorea_percentage_corona = calculate_percentage_in_corona(skorea_corona_stats)
```

Something that should be awkward to you is the number of cases in 80-89 year olds. 

Although they represent 8% of the population, about 18% of the cases is actually of that age group. Why? 

Unless there were major outbreaks in carehomes, it could be a bias in testing: Symptoms are more pronounced and therefore people get tested. If we assume that the distribution percentage is the same as the population, and that all the 80-89 year olds have shown symptopms AND have been tested, we can do an upper estimate of the total number of people that had the virus.
